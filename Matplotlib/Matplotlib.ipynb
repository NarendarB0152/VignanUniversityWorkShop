```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,48,53,54,48,43]
```

```python
plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(x,y,color="green",linewidth=5,linestyle="dotted")
```
